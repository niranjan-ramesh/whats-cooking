```python
import pandas as pd
import numpy as np
import ast
import re
import os
```

```python
def segregate(x):
    minutes = 0;
    x = x.strip()
    tokens = x.split(" ")
    for i in range(len(tokens)):
        token = tokens[i].strip()
        if i%2 != 0:
            if token == "hr":
                minutes += int(tokens[i-1].strip()) * 60
            elif token == "mins":
                minutes += int(tokens[i-1].strip())
    return str(minutes)


def process_time(input_df):
    time_taken = input_df.copy()
    time_taken['minutes'] = time_taken['minutes'].astype(str).apply(segregate).astype('int64')
    return time_taken
```

Read the recipe Datasets and display

```python
AllRecipes_recipes = pd.read_csv("temp/full_allrecipes.csv")
AllRecipes_recipes = process_time(AllRecipes_recipes)
display(AllRecipes_recipes.dtypes)
display(AllRecipes_recipes)
```

```python
DamnDelicious_recipes = pd.read_csv("temp/damn_delicious_recipes.csv")
DamnDelicious_recipes = DamnDelicious_recipes.drop(['ratings','ratings_count','tags','submitted'], axis=1)
DamnDelicious_recipes['contributor_id'] = "Unknown"
display(DamnDelicious_recipes.dtypes)
display(DamnDelicious_recipes)
```

```python
FoodCom_recipes = pd.read_csv("temp/full_food.csv")
FoodCom_recipes = FoodCom_recipes.rename(columns={"n_step":"n_steps"})
FoodCom_recipes['minutes'] = FoodCom_recipes['minutes'].fillna(0)
FoodCom_recipes['minutes'] = FoodCom_recipes['minutes'].astype(int)
display(FoodCom_recipes.dtypes)
display(FoodCom_recipes)
```

```python
FoodNetwork_recipes = pd.read_csv("temp/food_network_recipes.csv")
FoodNetwork_recipes = FoodNetwork_recipes.drop(['categories','servings'], axis=1)
FoodNetwork_recipes.columns = ['id','name','minutes','ingredients','steps','n_steps','n_ingredients']
FoodNetwork_recipes['contributor_id'] = "Unknown"
FoodNetwork_recipes['nutrition'] = "[]"
display(FoodNetwork_recipes.dtypes)
display(FoodNetwork_recipes)
```

To avoid recipe_id mismatch, merging the dataset with caution

```python
recipe_id_list = FoodCom_recipes['id'].tolist()
def get_count(num):
    for count in range(num, (FoodCom_recipes['id'].max())):
        if count not in recipe_id_list:
            recipe_id_list.append(count)
            return int(count);
```

```python
count = 0
for index, row in DamnDelicious_recipes.iterrows():
    count = get_count(count)
    DamnDelicious_recipes.at[index,'recipe_id'] = count
DamnDelicious_recipes['recipe_id'] = DamnDelicious_recipes['recipe_id'].astype("int64")
display(DamnDelicious_recipes)
```

```python
for index, row in AllRecipes_recipes.iterrows():
    count = get_count(count)
    AllRecipes_recipes.at[index,'recipe_id'] = count
AllRecipes_recipes['recipe_id'] = AllRecipes_recipes['recipe_id'].astype("int64")
display(AllRecipes_recipes)
```

```python
k = FoodCom_recipes['id'].max()+1
for index, row in FoodNetwork_recipes.iterrows():
    FoodNetwork_recipes.at[index,'recipe_id'] = k
    recipe_id_list.append(k)
    k = k+1
FoodNetwork_recipes['recipe_id'] = FoodNetwork_recipes['recipe_id'].astype("int64")
display(FoodNetwork_recipes)
```

## Adding All Four Datasets to One.

```python
full_recipes = pd.concat([FoodCom_recipes, DamnDelicious_recipes, AllRecipes_recipes, FoodNetwork_recipes], ignore_index=True)
full_recipes['id'] = full_recipes['recipe_id']
full_recipes = full_recipes.drop(['recipe_id'], axis=1)
full_recipes = full_recipes[['name', 'id', 'minutes', 'contributor_id', 'nutrition', 'n_steps', 'steps', 'ingredients', 'n_ingredients']]
```

```python
user_id_list = FoodCom_reviews['user_id'].to_list()
def get_next_idx(count):
    for i in range(count, (FoodCom_reviews['user_id'].max())):
        if i not in user_id_list:
            return i
```

```python
idx = 0

fn_unique_df = FoodNetwork_reviews['user_id'].unique()
fn_unique_df = pd.DataFrame(data=fn_unique_df, columns=['user_name'])

for index, row in fn_unique_df.iterrows():
    idx = get_next_idx(idx)
    fn_unique_df.at[index,'userid'] = idx
    user_id_list.append(idx)
fn_unique_df['userid'] = fn_unique_df['userid'].astype('int64')

FoodNetwork_reviews_copy = FoodNetwork_reviews.copy()
FoodNetwork_reviews_copy = FoodNetwork_reviews_copy[FoodNetwork_reviews_copy['user_id'].isnull() == False]
FoodNetwork_reviews_copy = FoodNetwork_reviews_copy[FoodNetwork_reviews_copy['user_id'] != "None"]
FoodNetwork_reviews_copy = FoodNetwork_reviews_copy[FoodNetwork_reviews_copy['review'].isnull() == False]

FoodNetwork_reviews_copy = pd.merge(left=FoodNetwork_reviews_copy, right=fn_unique_df, how="left", left_on="user_id", right_on="user_name")
FoodNetwork_reviews_copy['user_id'] = FoodNetwork_reviews_copy['userid']
FoodNetwork_reviews_copy['user_id'] = FoodNetwork_reviews_copy['user_id'].astype('int64')
FoodNetwork_reviews_copy = FoodNetwork_reviews_copy.drop(['userid','user_name'], axis=1)
FoodNetwork_reviews_copy.columns = ['user_id', 'recipeid', 'date', 'rating', 'review']

temp_df = FoodNetwork_recipes[['id', 'recipe_id']].copy()
temp_df = temp_df[temp_df['id'].str.isnumeric() == True]
temp_df['id'] = temp_df['id'].astype('float64').astype('int64')

FoodNetwork_reviews_final= pd.merge(left=FoodNetwork_reviews_copy, right=temp_df, how="left", left_on="recipeid", right_on="id")
FoodNetwork_reviews_final = FoodNetwork_reviews_final[['user_id','recipe_id','date','rating','review']]

display(FoodNetwork_reviews_final)
```
